# Test połączenia z serwerem prowadzącego

Ten notebook sprawdza czy możesz połączyć się z LLM-em hostowanym przez prowadzącego.

**Przed uruchomieniem:** upewnij się że jesteś w tej samej sieci WiFi co prowadzący.

```python
import requests
from openai import OpenAI

# === WPISZ ADRES PODANY PRZEZ PROWADZĄCEGO ===
INSTRUCTOR_SERVER = "http://192.168.0.142:1234"
# ==============================================

try:
    r = requests.get(f"{INSTRUCTOR_SERVER}/v1/models", timeout=3)
    models = [m["id"] for m in r.json().get("data", [])]
    if models:
        MODEL = models[0]
        print(f"✓ Połączenie OK! Dostępny model: {MODEL}")
    else:
        print("✗ Serwer odpowiada, ale brak załadowanego modelu. Powiedz prowadzącemu.")
        MODEL = None
except Exception as e:
    print(f"✗ Brak połączenia: {e}")
    print("  Sprawdź czy jesteś na tej samej sieci WiFi co prowadzący.")
    MODEL = None
```
